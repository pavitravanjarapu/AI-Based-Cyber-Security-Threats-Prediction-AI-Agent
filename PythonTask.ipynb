*Create a Python class BankAccount that allows users to deposit, withdraw, and check balance. Ensure balance doesn‚Äôt go below zero.*

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ‚Çπ{amount}. Current Balance: ‚Çπ{self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= amount
            print(f"Withdrawn ‚Çπ{amount}. Remaining Balance: ‚Çπ{self.balance}")

# Example
acc1 = BankAccount("Pavitra", 5000)
acc1.deposit(2000)
acc1.withdraw(1500)
acc1.withdraw(6000)
```

*Write a program that reads a text file and counts how many times each word appears.*

*Write a Python class Library to add, remove, and display available books*

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"‚úÖ '{book}' added to library")

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
            print(f"‚ùå '{book}' removed")
        else:
            print("Book not found!")

    def display_books(self):
        print("\nüìö Available Books:")
        for b in self.books:
            print("-", b)

# Example
lib = Library()
lib.add_book("Python Basics")
lib.add_book("Data Science 101")
lib.display_books()
lib.remove_book("Python Basics")
lib.display_books()
```

*Write a program that counts lines, words, and characters in a text file*

```python
with open("data.txt", "w") as f:
    f.write("Python is powerful.\nPython is easy to learn.\nKeep coding daily!")

lines = words = chars = 0
with open("data.txt", "r") as f:
    for line in f:
        lines += 1
        words += len(line.split())
        chars += len(line)

print(f"Lines: {lines}, Words: {words}, Characters: {chars}")
```

*Write a function to check whether a given string is a palindrome*

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

text = input("Enter a string: ")
print("‚úÖ Palindrome" if is_palindrome(text) else "‚ùå Not a palindrome")
```

*Use map, filter, and lambda to process a list of numbers*

```python
nums = [10, 25, 33, 50, 77, 82]

# Double each number
doubled = list(map(lambda x: x * 2, nums))

# Filter even numbers
even = list(filter(lambda x: x % 2 == 0, nums))

print("Original:", nums)
print("Doubled:", doubled)
print("Even Numbers:", even)
```

*Design an OOP-based calculator that performs +, -, , / operations*

```python
class Calculator:
    def add(self, a, b): return a + b
    def sub(self, a, b): return a - b
    def mul(self, a, b): return a * b
    def div(self, a, b):
        try:
            return a / b
        except ZeroDivisionError:
            return "Cannot divide by zero!"

calc = Calculator()
print("Addition:", calc.add(10, 5))
print("Subtraction:", calc.sub(10, 5))
print("Multiplication:", calc.mul(10, 5))
print("Division:", calc.div(10, 0))
```
